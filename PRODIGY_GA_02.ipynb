## üñºÔ∏è Generating an AI-Generated Image with OpenAI

### Steps:
1. üì¶ **Import Necessary Libraries**  
   - Loads OpenAI, requests, and PIL for handling image generation and display.

2. ü§ñ **Request Image Generation**  
   - Sends a prompt to OpenAI‚Äôs DALL¬∑E-2 model to generate an image of a futuristic city with flying cars.

3. üåê **Retrieve Image URL**  
   - The API returns a URL containing the generated image.

4. üîó **Display or Download the Image**  
   - Prints the image URL for access or further processing.


```python
import openai
import requests
from PIL import Image
from io import BytesIO

response = client.images.generate(
  model="dall-e-2",
  prompt="A futuristic city with flying cars",
  size="1024x1024",
  quality="standard",
  n=1,
)

image_url = response.data[0].url

print(image_url)
```

## üñºÔ∏è Displaying an AI-Generated Image

### Steps:
1. üì¶ **Import Required Libraries**  
   - Uses Matplotlib for visualization, PIL for image processing, and Requests for fetching the image.

2. üåê **Fetch Image from URL**  
   - Downloads the image from the provided URL.

3. üñºÔ∏è **Open the Image**  
   - Converts the downloaded content into an image format.

4. üìä **Display the Image**  
   - Uses Matplotlib to render and show the generated image.


```python
import matplotlib.pyplot as plt
from PIL import Image
import requests

def image_from_url(image_url):
    img = Image.open(requests.get(image_url, stream=True).raw)
    return img

img = image_from_url(image_url)
plt.imshow(img)
```
